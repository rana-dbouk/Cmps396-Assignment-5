```python
# @title Extract Keyframes Every N Seconds
import cv2
import os
import json

def extract_keyframes(video_path, interval_sec=4, output_dir="keyframes"):
    os.makedirs(output_dir, exist_ok=True)

    cap = cv2.VideoCapture(video_path)
    fps = cap.get(cv2.CAP_PROP_FPS)
    interval_frames = int(fps * interval_sec)

    count = 0
    saved = 0
    timestamps = []

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        if count % interval_frames == 0:
            timestamp = cap.get(cv2.CAP_PROP_POS_MSEC) / 1000.0  # Convert ms to sec
            filename = os.path.join(output_dir, f"frame_{saved:04d}.jpg")
            cv2.imwrite(filename, frame)
            timestamps.append({"filename": filename, "timestamp": timestamp})
            saved += 1

        count += 1

    cap.release()
    return timestamps

# Extract keyframes from downloaded video
keyframe_metadata = extract_keyframes("lecture_video.mp4", interval_sec=2)

# Save keyframe metadata
with open("image_metadata.json", "w") as f:
    json.dump(keyframe_metadata, f, indent=2)
```

##Generate and Save Image Embeddings Using CLIP

```python
import torch
from PIL import Image
from transformers import CLIPProcessor, CLIPModel
import json
import numpy as np

# Load keyframes
with open("image_metadata.json", "r") as f:
    keyframes = json.load(f)

# Load CLIP
clip_model = CLIPModel.from_pretrained("openai/clip-vit-base-patch32")
processor = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch32")

batch_size = 16
all_embeddings = []

for i in range(0, len(keyframes), batch_size):
    batch = keyframes[i:i+batch_size]
    images = [Image.open(f["filename"]).convert("RGB") for f in batch]

    inputs = processor(images=images, return_tensors="pt", padding=True)
    with torch.no_grad():
        emb = clip_model.get_image_features(**inputs)

    all_embeddings.append(emb.cpu().numpy())
    torch.cuda.empty_cache()  # Free up GPU memory if used

# Concatenate all batches
img_embeddings = np.vstack(all_embeddings)
np.save("image_embeddings.npy", img_embeddings)
```

##Create FAISS Index for Text Embeddings

```python
import faiss
import numpy as np
import json

# Load text embeddings and metadata
embeddings = np.load("text_embeddings.npy").astype('float32')
with open("text_metadata.json", "r") as f:
    metadata = json.load(f)

# Create FAISS index (Flat L2 index)
index = faiss.IndexFlatL2(embeddings.shape[1])  # dimension = embedding size (e.g., 384)
index.add(embeddings)  # Add all embeddings to the index
print("FAISS index contains", index.ntotal, "vectors.")
```

## Evaluate Semantic Search with FAISS Using Gold Dataset

```python
from sentence_transformers import SentenceTransformer
import numpy as np
import json

# Load model and data
model = SentenceTransformer("all-MiniLM-L6-v2")
SIMILARITY_THRESHOLD = 0.7

# Load FAISS index and metadata
# Assuming 'index' and 'metadata' are already loaded

# Load gold datasets
with open("gold_answerable.json", "r") as f:
    gold_answerable = json.load(f)

with open("gold_unanswerable.json", "r") as f:
    gold_unanswerable = json.load(f)

# Search function
def search(query, top_k=5, threshold=SIMILARITY_THRESHOLD):
    query_vec = model.encode([query], normalize_embeddings=True).astype('float32')
    D, I = index.search(query_vec, top_k)
    results = []
    for sim, idx in zip(D[0], I[0]):
        if sim >= threshold:
            results.append(metadata[idx])
    if not results:
        return ["Answer not found in the video."]
    return results

# Helper: Timestamp match
def is_timestamp_match(pred, gt_start, gt_end, tol=3.0):
    if isinstance(pred, str): return False
    return not (pred["end"] < gt_start - tol or pred["start"] > gt_end + tol)

# === Evaluation ===

correct = 0
total = len(gold_answerable)

print("🟩 Evaluating Answerable Questions:\n")
for item in gold_answerable:
    q = item["question"]
    gt_start = item["answer_start"]
    gt_end = item["answer_end"]
    results = search(q, top_k=5)
    matched = any(is_timestamp_match(r, gt_start, gt_end) for r in results)
    correct += matched
    print(f"Q: {q}\n✅ Matched: {matched}\n")

acc = correct / total
print(f"\n✔️ Answerable Accuracy: {acc:.2f} ({correct}/{total})\n")

# === Evaluate Unanswerable ===

false_positives = 0
print("🟥 Evaluating Unanswerable Questions:\n")
for q in gold_unanswerable:
    results = search(q, top_k=5)
    returned_irrelevant = any(isinstance(r, dict) for r in results)
    false_positives += returned_irrelevant
    print(f"Q: {q}\n❌ Returned Irrelevant Result: {returned_irrelevant}\n")

rej_rate = 1 - (false_positives / len(gold_unanswerable))
print(f"\n🚫 Unanswerable Rejection Rate: {rej_rate:.2f} ({len(gold_unanswerable) - false_positives}/{len(gold_unanswerable)})")
```

##Prepare Corpus and Timestamps for Lexical Retrieval (TF-IDF & BM25)

```python
import json

# Load text segments
with open("cleaned_segments.json", "r") as f:
    transcript_segments = json.load(f)

corpus = [seg["text"] for seg in transcript_segments]
timestamps = [(seg["start"], seg["end"]) for seg in transcript_segments]
```

## Define TF-IDF Search Over Transcript Segments

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Build TF-IDF matrix
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(corpus)

def tfidf_search(query, top_k=5):
    query_vec = tfidf_vectorizer.transform([query])
    similarities = cosine_similarity(query_vec, tfidf_matrix).flatten()
    top_indices = np.argsort(similarities)[::-1][:top_k]

    results = []
    for i in top_indices:
        results.append({
            "start": timestamps[i][0],
            "end": timestamps[i][1],
            "text": corpus[i],
            "score": similarities[i]
        })
    return results
```

##Define BM25 Search Over Transcript Segments

```python
from rank_bm25 import BM25Okapi

# Tokenize corpus for BM25
tokenized_corpus = [doc.lower().split() for doc in corpus]
bm25 = BM25Okapi(tokenized_corpus)

def bm25_search(query, top_k=5):
    tokenized_query = query.lower().split()
    scores = bm25.get_scores(tokenized_query)
    top_indices = np.argsort(scores)[::-1][:top_k]

    results = []
    for i in top_indices:
        results.append({
            "start": timestamps[i][0],
            "end": timestamps[i][1],
            "text": corpus[i],
            "score": scores[i]
        })
    return results
```

##Evaluate TF-IDF or BM25 Retrieval Using Gold Dataset

```python
def is_timestamp_match(pred, gt_start, gt_end, tol=3.0):
    return not (pred["end"] < gt_start - tol or pred["start"] > gt_end + tol)

def evaluate_retrieval(search_func, name="TF-IDF"):
    print(f"\n🔎 Evaluating {name} Retrieval")

    # Answerable
    correct = 0
    for item in gold_answerable:
        q = item["question"]
        gt_start, gt_end = item["answer_start"], item["answer_end"]
        preds = search_func(q, top_k=5)
        matched = any(is_timestamp_match(p, gt_start, gt_end) for p in preds)
        correct += matched
        print(f"Q: {q}\n✅ Matched: {matched}\n")

    acc = correct / len(gold_answerable)
    print(f"✔️ {name} Answerable Accuracy: {acc:.2f} ({correct}/{len(gold_answerable)})")

    # Unanswerable
    false_positives = 0
    for q in gold_unanswerable:
        preds = search_func(q, top_k=5)
        if any(isinstance(p, dict) for p in preds):
            false_positives += 1

    rej_rate = 1 - false_positives / len(gold_unanswerable)
    print(f"🚫 {name} Unanswerable Rejection Rate: {rej_rate:.2f} ({len(gold_unanswerable)-false_positives}/{len(gold_unanswerable)})")
```

```python
evaluate_retrieval(tfidf_search, name="TF-IDF")
evaluate_retrieval(bm25_search, name="BM25")
```
